```python
import sys
sys.path.insert(0, "../src")

import augment

import numpy as np
from astropy import constants as const
from astropy import units as u
```

```python
sampler = augment.Sampler(n_samples=100_000, seed=42)

sampler.add(
    "T_1",
    5772*u.K,
    upper_error=800*u.K,
    lower_error=600*u.K,
    lower=0*u.K,
    dist="split_normal",
)

sampler.add(
    "T_2",
    1702.0*u.K,
    upper_error=200*u.K,
    lower_error=200*u.K,
    lower=0*u.K,
    dist="split_normal",
)

def operators(temperature_1, temperature_2):
    return {
        "addition": temperature_1 + temperature_2,
        "subtraction": temperature_1 - temperature_2,
        "multiplication": temperature_1 * temperature_2,
        "division": temperature_1 / temperature_2,
        "power": np.sqrt(temperature_1 ** 2 - temperature_2 ** 2),
    }

out = sampler.run(
    operators,
    return_samples=True,
    error_budget_against="power",
)

print(augment.format_summary_table(
    out,
    keys=("addition", "subtraction", "multiplication", "division", "power"),
    tablefmt="plain",
))

fig = augment.plot_corner(
    out,
    keys=("addition", "subtraction", "multiplication", "division", "power"),
    title_fmt=".2e",
)
```

```python
import numpy as np
from astropy import units as u
from uncertainties import ufloat
from uncertainties import umath

# --- augment result ---
sampler = augment.Sampler(n_samples=100_000, seed=42)

sampler.add(
    "T_1",
    5772 * u.K,
    upper_error=800 * u.K,
    lower_error=600 * u.K,
    lower=0 * u.K,
    dist="split_normal",
)

sampler.add(
    "T_2",
    1702.0 * u.K,
    upper_error=200 * u.K,
    lower_error=200 * u.K,
    lower=0 * u.K,
    dist="split_normal",
)

def operators(T_1, T_2):
    return {
        "addition": T_1 + T_2,
        "subtraction": T_1 - T_2,
        "multiplication": T_1 * T_2,
        "division": T_1 / T_2,
        "power": np.sqrt(T_1 ** 2 - T_2 ** 2),
    }

out = sampler.run(
    operators,
    return_samples=True,
    error_budget_against="power",
)

keys = ("addition", "subtraction", "multiplication", "division", "power")

# --- uncertainties result ---
T1_u = ufloat(5772.0, 0.5 * (800.0 + 600.0))  # symmetric approximation
T2_u = ufloat(1702.0, 200.0)

unc = {
    "addition": T1_u + T2_u,
    "subtraction": T1_u - T2_u,
    "multiplication": T1_u * T2_u,
    "division": T1_u / T2_u,
    "power": umath.sqrt(T1_u**2 - T2_u**2),
}

units = {
    "addition": "K",
    "subtraction": "K",
    "multiplication": "K2",
    "division": "dimensionless",
    "power": "K",
}


# --- comparison table ---
print()
print("Comparison to uncertainties package")
print(
    f"{'Parameter':<15} "
    f"{'aug median':>15} "
    f"{'aug +err':>15} "
    f"{'aug -err':>15} "
    f"{'aug mean err':>15} "
    # f"{'aug std':>15} "
    f"{'unc value':>15} "
    f"{'unc error':>15} "
    f"{'Unit':>15}"
)
print("-" * 128)

for key in keys:
    sample = out["samples"][key]
    vals = sample.to_value(sample.unit)
    vals = vals[np.isfinite(vals)]

    summary = out["summary"][key]
    aug_med = summary["median"].to_value(sample.unit)
    aug_plus = summary["plus"].to_value(sample.unit)
    aug_minus = summary["minus"].to_value(sample.unit)
    aug_mean_err = 0.5 * (aug_plus + aug_minus)
    aug_std = np.std(vals, ddof=1)

    unc_val = unc[key].nominal_value
    unc_sig = unc[key].std_dev

    print(
        f"{key:<15} "
        f"{aug_med:15.6e} "
        f"{aug_plus:15.6e} "
        f"{aug_minus:15.6e} "
        f"{aug_mean_err:15.6e} "
        # f"{aug_std:15.6e} "
        f"{unc_val:15.6e} "
        f"{unc_sig:15.6e} "
        f"{units[key]:>15}"
    )
```

```python
import numpy as np
from astropy import units as u
from uncertainties import ufloat
from uncertainties import umath

# --- augment result ---
sampler = augment.Sampler(n_samples=100_000, seed=42)

sampler.add(
    "x_1",
    32.7 * u.dimensionless_unscaled,
    upper_error=2.3 * u.dimensionless_unscaled,
    lower_error=2.1 * u.dimensionless_unscaled,
    dist="split_normal",
)

sampler.add(
    "x_2",
    17.3 * u.dimensionless_unscaled,
    upper_error=3.4 * u.dimensionless_unscaled,
    lower_error=2.8 * u.dimensionless_unscaled,
    dist="split_normal",
)

sampler.add(
    "theta",
    45.2 * u.deg,
    upper_error=7.6 * u.deg,
    lower_error=7.6 * u.deg,
    dist="split_normal",
)

def numpy_functions(x_1, x_2, theta):
    return {
        "np.sin": np.sin(theta),
        "np.cos": np.cos(theta),
        "np.exp": np.exp(x_1),
        "np.log": np.log(x_1),
        "np.power": np.power(x_1, x_2),
        "np.sqrt": np.sqrt(x_1),
        "np.cbrt": np.cbrt(x_1),
    }

out = sampler.run(
    numpy_functions,
    return_samples=True,
    error_budget_against="np.cbrt",
)

keys = ("np.sin", "np.cos", "np.exp", "np.log", "np.power", "np.sqrt", "np.cbrt")
log_keys = keys

print(augment.format_summary_table(
    out,
    keys=keys,
    log_keys=log_keys,
    tablefmt="plain",
))


# --- uncertainties result ---
x1_u = ufloat(32.7, 0.5 * (2.3 + 2.1))
x2_u = ufloat(17.3, 0.5 * (3.4 + 2.8))
theta_u = ufloat(np.deg2rad(45.2), np.deg2rad(7.6))

unc = {
    "np.sin": umath.sin(theta_u),
    "np.cos": umath.cos(theta_u),
    "np.exp": umath.exp(x1_u),
    "np.log": umath.log(x1_u),
    "np.power": x1_u ** x2_u,
    "np.sqrt": umath.sqrt(x1_u),
    "np.cbrt": x1_u ** (1.0 / 3.0),
}

units = {
    "np.sin": "dimensionless",
    "np.cos": "dimensionless",
    "np.exp": "dimensionless",
    "np.log": "dimensionless",
    "np.power": "dimensionless",
    "np.sqrt": "dimensionless",
    "np.cbrt": "dimensionless",
}

# --- comparison table ---
print()
print("Comparison to uncertainties package")
print(
    f"{'Parameter':<12} "
    f"{'aug median':>14} "
    f"{'aug +err':>14} "
    f"{'aug -err':>14} "
    f"{'aug mean err':>14} "
    # f"{'aug std':>14} "
    f"{'unc value':>14} "
    f"{'unc sigma':>14}"
    f"{'Unit':>15}"
)
print("-" * 118)

for key in keys:
    sample = out["samples"][key]
    vals = sample.to_value(sample.unit)
    vals = vals[np.isfinite(vals)]

    s = out["summary"][key]
    aug_med = s["median"].to_value(sample.unit)
    aug_plus = s["plus"].to_value(sample.unit)
    aug_minus = s["minus"].to_value(sample.unit)
    aug_mean_err = 0.5 * (aug_plus + aug_minus)
    aug_std = np.std(vals, ddof=1)

    unc_val = unc[key].nominal_value
    unc_sig = unc[key].std_dev

    print(
        f"{key:<12} "
        f"{aug_med:14.6e} "
        f"{aug_plus:14.6e} "
        f"{aug_minus:14.6e} "
        f"{aug_mean_err:14.6e} "
        # f"{aug_std:14.6e} "
        f"{unc_val:14.6e} "
        f"{unc_sig:14.6e}"
        f"{units[key]:>15}"
    )
```
